```python
from geopy.geocoders import Nominatim
import sys
sys.path.append('/home/ilke/Desktop/Coursera_Capstone')
from config import CLIENT_ID, CLIENT_SECRET, VERSION
import requests
import pandas as pd
```

```python
df = pd.read_csv("city_list.csv", header=None)
df.columns = ["city", "population", "area(km2)"]
df.population = df.population.astype(int)
df.head()
```

```python
def get_vegan_venue_count(city):
    CITY = city
    search_query = "Vegan"
    url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&near={}&v={}&query={}&limit=50'.format(CLIENT_ID, CLIENT_SECRET, CITY, VERSION, search_query)
    results = requests.get(url).json()
    return len(results['response']['venues'])

def get_vegatarian_venue_count(city):
    CITY = city
    search_query = "Vegetarian"
    url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&near={}&v={}&query={}&limit=50'.format(CLIENT_ID, CLIENT_SECRET, CITY, VERSION, search_query)
    results = requests.get(url).json()
    return len(results['response']['venues'])
```

```python
df["vegan_venue_count"] = df.city.apply(get_vegan_venue_count)
```

```python
df["vegetarian_venue_count"] = df.city.apply(get_vegatarian_venue_count)
```
